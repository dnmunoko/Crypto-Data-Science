- Helps solve a wide range of business problems.
- Provides best-in-class accuracy.
- Power:

    - Produces state-of-the-art results without extensive data training typically required by ML.
- Boost:
    - Based on gradient boosting library.
    - Gradient Boosting:
        - Powerful ML algorithm.
        - Widely applied to business: fraud detection, recommendation items, forecasting, predictions, classifier, etc.
        - Can return very good results with relatively less data unlike deep learning models that need a lot of data. 

### What is Gradient Boosting?
- ML technique for regression and classification problems.
- Result:
    - Prediction model in the form of an ensemble weak prediction models. 
    - Typically made of decision trees.
- Decision Tree Model
    - Series of Yes/No rules based on features to match all possible outcomes of a decision.
- Decision Tree
    - Tree shaped diagram.
    - Determines a course of action, category.
    - Each branch represents a possible decision, occurence, or reaction.
    - Pros:
        - Easy to understand.
    - Cons:
        - Too simple
        - Not powerful enough for complex data
        - Not often used on its own for prediction
- Random Forest:
    - Takes advantage of many decision trees.
    - Rules created from subsamples of features.
    - Each tree is weaker than a full decision tree.
    - Combining trees produces better overall performance.
    - Pros:
        - Wisdom of the crowd
        - Results in high quality models
        - Fast to train
    - Cons:
        - Model can get huge
        - Hard to understand predictions
- Gradient Boosting Model
    - Uses even weaker decision trees.
    - Increasingly focused on hard examples.
    - Pros:
        - High performing
    - Cons:
        - Small change in feature set or training set can radically change model.
        - Hard to understand predictions.
- Gradient Boosting
    - Boostimg method
    - Ensemble learning technique
    - Combines models
- Summary:
    - Combines the predictions from 10s, 100s, or even 1000s of tree models.
    - Iteratively learns from weak learners to build a strong model.
    - Optimizes regression, classification, and ranking.

### Load Data

#### Question: Whether We Should Buy or Sell Crypto

```python
#!pip install yfinance
```

```python
import yfinance

symbol = "XRP-USD"

start = "2014-01-01"

end = "2021-01-01"

dataframe = yfinance.download(symbol, start, end)

dataframe
```

```python
dataframe.to_csv("xrp_2014_2021.csv")
```

---

### Process Data

```python
#Create a column that can be used for classification: wheter buy or sell crypto
import numpy as np

BUY = 1

SELL = 0

dataframe["Buy/Sell"] = np.where(dataframe["Close"].shift(-1) > dataframe["Close"],
                                BUY,
                                SELL)

dataframe
```

```python
# Define y and X
y = dataframe["Buy/Sell"].values

y
```

```python
X = dataframe[["High", "Low", "Close", "Volume"]].values

X
```

```python
#Split data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#Build CatBoost Classifier Model
#!pip install catboost
```

```python
import catboost

from catboost import CatBoostClassifier

model = CatBoostClassifier(iterations = 10, learning_rate = 0.1)

model.fit(X_train, y_train)

model.is_fitted()
```

```python
models_predictions = model.predict(X_test)

models_predictions
```

```python
#Accuracy
model.score(X_test, y_test)
```
