# Radius Neighbors

- Used for classification and regression.
- It's an extension of k-NN.
- How are they similar? They store entire training set.
- How are they different?
    - How you use the training dataset during prediction.
    - RN makes predictions using all samples in the radius of a sample rather than using the k-NN.
- When is it used?
    - More for sparse (scattered) data.
    - When we need contribution of neighbors to be proportional to the density of samples.
- How does it work?
    - Dense regions will contribute more information.
    - Sparse regions will contribute less information.
- Why?
    - Prevent samples far away from contributing to a prediction.
- Radius:
    - Assumes the input variables are numeric and scaled to the range 0-1, e.g. normalized.
- Curse of Dimensionality:
    - If radius is fixed, it is less effective as the number of features increases this causes the samples to spread further apart.

### Load Stock Data With Yahoo API Finance

```python
import yfinance

symbol = "BTC-USD"

start = "2015-01-01"

end = "2021-01-01"

dataframe = yfinance.download(symbol, start, end)

dataframe
```

```python
dataframe.to_csv("btc-usd-2014-2021.csv")
```

```python
#Add new feature
import numpy as np

BUY = 1

SELL = 0

dataframe["Buy or Sell (Adj Close)"] = np.where(dataframe["Adj Close"].shift(-1) > dataframe["Adj Close"],
                                               BUY,
                                               SELL)

dataframe
```
